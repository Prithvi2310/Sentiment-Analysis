```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
#reading csv file
data = pd.read_csv('.Data/Sumarized.csv')
data.head()
```

```python
#converting cleaned data to list
data_list = []

for i in range(len(data)):
    data_list.append(data.loc[i,"Summary"])
```

```python
#performing sentiment-analysis
result = []

for i in data_list:
  result.append(classifier(i))
```

```python
#extracting the sentiment from classifiers output
final_sentiment = []

for i in range(len(result)):
  final_sentiment.append(result[i][0]['label'])
```

```python
#merging it in the dataframe
data['Tranformer_sentiment'] = final_sentiment
data.head()
```
